```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
fig, axs = plt.subplots(2,2) # axs : multiple axes
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0,0].plot(x, y)
axs[1, 0].scatter(x, y)
axs
```
